## Libraries

```python
import os
os.chdir("D:/Machine Learning/")

import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder

from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression,Ridge,Lasso

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

## Preprocessing

```python
x=df.drop(['Price','id','Model','Min.Price','Make'],axis=1)
y=df['Price']
```

```python
cat=x.select_dtypes(include='object')
cat
```

```python
con=x.select_dtypes(exclude='object')
con
```

### Categorical Data Preprocessing

```python
cat.isna().sum()
```

```python
cat['AirBags']=cat['AirBags'].fillna(cat['AirBags'].mode().iloc[0])
```

```python
cat.isna().sum()
```

```python
la=LabelEncoder()
```

```python
cat1=pd.DataFrame(cat.apply(la.fit_transform))
cat1
```

### Numrical Data Preprocessing

```python
con.isna().sum()
```

```python
con['Rear.seat.room']=con['Rear.seat.room'].fillna(con['Rear.seat.room'].median())
con['Luggage.room']=con['Luggage.room'].fillna(con['Luggage.room'].median())
```

```python
con.isna().sum()
```

```python
ss=StandardScaler()
```

```python
con1=pd.DataFrame(ss.fit_transform(con),columns=ss.get_feature_names_out())
con1
```

```python
x1=con1.join(cat1)
x1
```

## Feature Selection

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
lr=LinearRegression()
```

```python
sfs=SequentialFeatureSelector(lr)
```

```python
sfs.fit(x1,y)
```

```python
x2=pd.DataFrame(x,columns=sfs.get_feature_names_out())
x2
```

### Preprocessing

```python
cat=[]
con=[]

for i in x2.columns:
    if x[i].dtypes=='object':
        cat.append(i)
    else:
        con.append(i)
```

```python
cat
```

```python
con
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy=('median'))),('scalar',StandardScaler())])
cat_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='most_frequent')),('encoder',OneHotEncoder())])
```

```python
pre=ColumnTransformer([('con_pipe',num_pipe,con),('cat_pipe',cat_pipe,cat)])
pre
```

```python
x3=pd.DataFrame(pre.fit_transform(x2),columns=pre.get_feature_names_out())
x3
```

### Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x3,y,test_size=0.2,random_state=23)
```

### Model Building

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

## Traning Data Evaluation

```python
y_pred_train=lr.predict(x_train)

mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
r=r2_score(y_pred_train,y_train)
rmse=mse**0.5

print('MSE:',mse)
print('MAE:',mae)
print('R2:',r)
print('RMSE:',rmse)
```

## Testing Data Evaluation

```python
y_pred_test=lr.predict(x_test)

mse1=mean_squared_error(y_pred_test,y_test)
mae1=mean_absolute_error(y_pred_test,y_test)
r1=r2_score(y_pred_test,y_test)
rmse1=mse**0.5

print('MSE:',mse1)
print('MAE:',mae1)
print('R2:',r1)
print('RMSE:',rmse1)
```
