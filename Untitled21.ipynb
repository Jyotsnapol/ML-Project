## Linear Regression

```python
import pandas as pd
```

```python
df=pd.read_csv("D:\Basic Python\Cars93.csv")
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

## Preprocessing

```python
x
```

```python
cat= x.select_dtypes(include='object')
cat
```

```python
con= x.select_dtypes(exclude='object')
con
```

## catagorical data Preprocessing

```python
cat
```

```python
cat.isna().sum()
```

```python
cat['AirBags']= cat['AirBags'].fillna(cat['AirBags'].mode().iloc[0])
```

```python
cat.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
la=LabelEncoder()
oh=OneHotEncoder()
```

```python
cat1=pd.DataFrame(cat.apply(la.fit_transform))                      #label Encoding
cat1
```

```python
cat2= pd.DataFrame(oh.fit_transform(cat).toarray(),columns=oh.get_feature_names_out())
cat2
```

## Numrical Data Preprocessing

```python
con
```

```python
con.isna().sum()
```

```python
con['Rear.seat.room']=con['Rear.seat.room'].fillna(con['Rear.seat.room'].median())
con['Luggage.room']=con['Luggage.room'].fillna(con['Luggage.room'].median())
```

```python
con.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
mn=MinMaxScaler()
ss=StandardScaler()
```

```python
con1=pd.DataFrame(mn.fit_transform(con),columns=mn.get_feature_names_out())
con1
```

```python
con2=pd.DataFrame(ss.fit_transform(con),columns=ss.get_feature_names_out())
con2
```

## Final Dtaframe

```python
x1= cat1.join(con1)
x1
```

## Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
sfs=SequentialFeatureSelector(lr,n_features_to_select=17,direction='forward')
```

```python
x2=pd.DataFrame(sfs.fit_transform(x1,y),columns=sfs.get_feature_names_out())
x2
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x2,y,test_size=0.2,random_state=23)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

## Evaluation Of Model

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

## Traning Data Evaluation

```python
y_pred_train=lr.predict(x_train)
y_pred_train
```

```python
y_train
```

```python
mse=mean_squared_error(y_pred_train,y_train)
mae=mean_absolute_error(y_pred_train,y_train)
r=r2_score(y_pred_train,y_train)
rmse=mse**0.5

print('MSE:',mse)
print('RMSE:',rmse)
print('MAE:',mae)
print('R2:',r)
```

## Testing Data Evaluation

```python
y_pred_test=lr.predict(x_test)
y_pred_test
```

```python
y_test
```

```python
mse1=mean_squared_error(y_pred_test,y_test)
mae1=mean_absolute_error(y_pred_test,y_test)
r1=r2_score(y_pred_test,y_test)
rmse1=mse1**0.5

print('MSE:',mse1)
print('RMSE:',rmse1)
print('MAE:',mae1)
print('R2:',r1)
```
