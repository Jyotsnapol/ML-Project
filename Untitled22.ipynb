## Logistic Regression

### Data Collection

```python
import pandas as pd
```

```python
df=pd.read_csv("D:\Machine Learning\Cars94.csv")
df
```

### EDA

```python
df.info()
```

```python
x=df.drop(['Man.trans.avail'],axis=1)
y=df['Man.trans.avail']
```

### Preprocessing

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
ss=StandardScaler()
la=LabelEncoder()
```

```python
x1=pd.DataFrame(ss.fit_transform(x),columns=ss.get_feature_names_out())
x1
```

```python
y1=la.fit_transform(y)
y1
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=23)
```

### Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

### Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

### Traning Data Evaluation

```python
y_pred_train=lr.predict(x_train)

acc=accuracy_score(y_pred_train,y_train)
clr=classification_report(y_pred_train,y_train)
cnf=confusion_matrix(y_pred_train,y_train)

print('accuracy_score:',acc)
print('classification_report:\n',clr)
print('confusion_matrix:\n',cnf)
```

### Testing Data Evaluation

```python
y_pred_test=lr.predict(x_test)

acc1=accuracy_score(y_pred_test,y_test)
clr1=classification_report(y_pred_test,y_test)
cnf1=confusion_matrix(y_pred_test,y_test)

print('accuracy_score:',acc1)
print('classification_report:\n',clr1)
print('confusion_matrix:\n',cnf1)
```
